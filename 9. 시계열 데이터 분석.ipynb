```python
# 라이브러리 로드
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# os에 따라 시각화 폰트 설정하기
import os

if os.name == 'nt':
    font_family = "Malgun Gothic"
else:
    font_family = "AppleGothic"
    
sns.set(font=font_family, rc={"axes.unicode_minus" : False })
```

```python
# retina 를 지원하는 디스플레이에서 폰트가 선명하게 보이기 위함
from IPython.display import set_matplotlib_formats

set_matplotlib_formats("retina")
```

```python
# csv 파일 로드
df = pd.read_csv("bd_kma_weather.csv")
df.shape
```

```python
# 데이터를 로드해 오면 object 타입입니다.
df["일시"].head(1)
```

```python
# datetime 형식으로 변경해 줍니다.
df["일시"] = pd.to_datetime(df["일시"])
```

```python
# datetime 형식으로 변경이 되었는지 확인합니다.
df["일시"].head(1)
```

### 파생변수 만들기

```python
# 연, 월, 일 파생변수를 생성합니다.
df["연"] = df["일시"].dt.year
df["월"] = df["일시"].dt.month
df["일"] = df["일시"].dt.day
```

```python
df[["일시", "연", "월", "일"]].head()
```

### 시계열 데이터 시각화

```python
plt.figure(figsize=(15, 4))
plt.title("일자별 평균 기온")
sns.lineplot(data=df, x="일시", y="평균기온(°C)", ci=None)
```

```python
plt.figure(figsize=(15, 4))
plt.title("연 평균 기온")
sns.pointplot(data=df, x="연", y="평균기온(°C)", ci=None)
```

```python
plt.figure(figsize=(15, 4))
plt.title("월별 평균 기온")
sns.pointplot(data=df, x="월", y="평균기온(°C)", ci=None)
```

```python
plt.figure(figsize=(15, 4))
plt.title("월 지점별 평균 기온")
sns.pointplot(data=df, x="월", y="평균기온(°C)", ci=None, hue="지점명")
```
